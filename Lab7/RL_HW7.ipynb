```python
import gym
from gym.wrappers import RecordVideo
import numpy as np
import os
import glob
import io
import base64
from IPython.display import HTML
from IPython import display as ipythondisplay
import matplotlib.pyplot as plt

# Fourier Basis
class FourierBasis:
    def __init__(self, state_dim, order):
        self.state_dim = state_dim
        self.order = order
        self.coefficients = np.indices([order + 1] * state_dim).reshape(state_dim, -1).T

    def encode(self, state):
        scaled_state = (state - np.min(state)) / (np.max(state) - np.min(state))
        return np.cos(np.pi * np.dot(self.coefficients, scaled_state))
#RBF Encoder
class RBFEncoder:
    def __init__(self, state_dim, num_centers, width):
        self.num_centers = num_centers
        self.width = width
        # Generate centers uniformly in the state space
        self.centers = np.random.uniform(-1, 1, (num_centers, state_dim))

    def encode(self, state):
        # Compute RBF features
        return np.exp(-np.linalg.norm(self.centers - state, axis=1)**2 / (2 * self.width**2))
#Tile Encoder
class TileCoder:
    def __init__(self, state_dim, num_tilings, tiles_per_dim, low, high):
        self.num_tilings = num_tilings
        self.tiles_per_dim = tiles_per_dim
        self.state_dim = state_dim
        self.low = low
        self.high = high

        # Compute the width of each tile
        self.tile_width = (high - low) / (tiles_per_dim - 1)

        # Compute offsets for each tiling
        self.offsets = [(i / num_tilings) * self.tile_width for i in range(num_tilings)]

    def encode(self, state):
        """
        Encode the continuous state into tile-coded features.
        """
        scaled_state = (state - self.low) / (self.high - self.low)  # Normalize state
        features = []
        for offset in self.offsets:
            feature = np.floor((scaled_state + offset) / self.tile_width)
            features.append(feature)
        return np.concatenate(features).astype(int)
```

```python
from gym.wrappers import TimeLimit
from gym.wrappers import StepAPICompatibility

# Video display
def show_video(video_folder='./video'):
    mp4list = glob.glob(f'{video_folder}/*.mp4')
    if len(mp4list) > 0:
        mp4 = mp4list[-1]
        video = io.open(mp4, 'r+b').read()
        encoded = base64.b64encode(video)
        display(HTML(data=f'''
            <video alt="test" autoplay loop controls style="height: 400px;">
            <source src="data:video/mp4;base64,{encoded.decode('ascii')}" type="video/mp4" />
            </video>'''))
    else:
        print("Could not find video")

# Wrap environment for video recording
def wrap_env(env, video_folder='./video', run_id=0, env_name=''):
    video_folder = f"{video_folder}/{env_name}/run_{run_id}"
    if not os.path.exists(video_folder):
        os.makedirs(video_folder)

    return RecordVideo(env, video_folder, episode_trigger=lambda episode_id: True)

#Monte Carlo Simulation
def monte_carlo(
    env_name,
    episodes=100,
    gamma=0.9,
    alpha=0.05,
    epsilon=0.3,
    epsilon_min=0.05,
    decay_rate=0.999,
    num_centers=150,
    width=0.05,
    reward_modifier=None,  # Optional reward adjustment to change rewards
    max_step=500,
):
    env = gym.make(env_name, render_mode="rgb_array", new_step_api=True)
    state_dim = env.observation_space.shape[0]
    action_dim = env.action_space.n

    rbf = RBFEncoder(state_dim, num_centers, width)
    weights = np.random.normal(0, 0.1, (action_dim, num_centers))

    rewards_per_episode = []

    for ep in range(episodes):
        state = env.reset(seed=42)
        done = False
        trajectory = []
        total_reward = 0

        while not done:
            # Epsilon-greedy policy
            if np.random.rand() < epsilon:
                action = env.action_space.sample()
            else:
                features = rbf.encode(state)
                q_values = np.dot(weights, features)
                action = np.argmax(q_values)

            # Step the environment
            step_result = env.step(action)
            next_state, reward, terminated, truncated, _ = step_result if len(step_result) == 5 else (*step_result[:3], None)
            done = terminated or truncated

            # Custom reward modification
            if reward_modifier is not None:

                if env_name == "MountainCar-v0":
                    reward = mountaincar_reward_modifier(state, action, reward, done)
                elif env_name == "CartPole-v1":
                    reward = cartpole_reward_modifier(state, action, reward, max_step)

            # Default reward update without mods
            total_reward += reward
            trajectory.append((state, action, reward))
            state = next_state

            # Termination
            done = terminated or truncated

        rewards_per_episode.append(total_reward)

        # Monte Carlo updates
        G = 0
        mean_reward = np.mean(rewards_per_episode[-100:]) if len(rewards_per_episode) >= 100 else 0
        std_reward = np.std(rewards_per_episode[-100:]) if len(rewards_per_episode) >= 100 else 1e-3
        for state, action, reward in reversed(trajectory):
            G = reward + gamma * G
            #G = np.clip((G - mean_reward) / std_reward, -10, 10) Normalization
            features = rbf.encode(state)
            q_value = np.dot(weights[action], features)
            weights[action] += alpha * (G - q_value) * features

        epsilon = max(epsilon_min, epsilon * decay_rate)

        if ep % 10 == 0:
            avg_reward = np.mean(rewards_per_episode[-10:])
            print(f"Episode {ep}, Avg Reward: {avg_reward:.2f}, Epsilon: {epsilon:.4f}")

    return weights, rewards_per_episode, rbf  # Return weights and encoder



def playback_episode(env_name, trained_weights, rbf, num_centers,max_step):
    # Wrap the environment for playback
    env = wrap_env(gym.make(env_name, render_mode="rgb_array", new_step_api=True),
                   video_folder='./video',
                   run_id=3,
                   env_name=env_name)
    env = TimeLimit(env, max_episode_steps=500)

    # Reset environment
    state = env.reset(seed=42)

    done = False
    step_count = 0
    total_reward = 0
    max_steps = max_step  # Maximum steps for playback

    print("Starting playback of an episode...")

    while step_count < max_steps and not done:
        # Compute action using the trained weights
        features = rbf.encode(state)
        q_values = np.dot(trained_weights, features)
        action = np.argmax(q_values)

        step_result = env.step(action)
        if len(step_result) == 5:
            state, reward, terminated, truncated, info = step_result
            done = terminated or truncated
        else:
            state, reward, done, info = step_result

        total_reward += reward
        step_count += 1
        print(f"Step {step_count}: State={state}, Action={action}, Reward={reward}, Total={total_reward}")

    print(f"Playback complete. Total reward for this episode: {total_reward:.2f}")

    env.close()
    show_video(f'./video/{env_name}/run_3')

def cartpole_reward_modifier(state, action, reward, step_count):


    # Scaling survival reward
    reward += step_count * 0.1  # Incremental reward based on steps survived


    return reward



def mountaincar_reward_modifier(state, action, reward, done):
    # Extract position and velocity
    position = state[0]
    velocity = state[1]

    # Scale reward continuously based on position
    normalized_position = (position + 1.2) / 1.7  # Scale to range [0, 1]
    reward += 10 * normalized_position  # Scale reward proportionally to position

    # Reward moving to the right
    if velocity > 0:  # Positive velocity indicates moving right
        reward += 5 * velocity  # Reward scaled by velocity magnitude

    # Penalize moving left
    if velocity < 0:
        reward -= 2 * abs(velocity)  # Penalty scaled by velocity magnitude

    # Success boost if the goal is reached
    if done and position >= 0.5:  # Goal state in MountainCar
        reward += 1000  # Large bonus for success

    return reward



```

```python
def semi_gradient_sarsa(
    env_name,
    episodes=200,
    gamma=0.9,
    alpha=0.05,
    epsilon=0.3,
    epsilon_min=0.05,
    decay_rate=0.9995,
    num_centers=150,
    width=0.05,
    reward_modifier=None  # Optional reward adjustment function
):
    env = gym.make(env_name, render_mode="rgb_array", new_step_api=True)
    state_dim = env.observation_space.shape[0]
    action_dim = env.action_space.n

    rbf = RBFEncoder(state_dim, num_centers, width)
    weights = np.random.normal(0, 0.1, (action_dim, num_centers))

    rewards_per_episode = []

    for ep in range(episodes):
        # Initialize the state
        reset_result = env.reset(seed=42)
        state = reset_result[0] if isinstance(reset_result, tuple) else reset_result
        features = rbf.encode(state)

        # Choose initial action using epsilon-greedy policy
        if np.random.rand() < epsilon:
            action = env.action_space.sample()
        else:
            q_values = np.dot(weights, features)
            action = np.argmax(q_values)

        done = False
        total_reward = 0

        while not done:
            # Take a step in the environment
            step_result = env.step(action)
            next_state, reward, terminated, truncated, _ = step_result if len(step_result) == 5 else (*step_result[:3], None)
            done = terminated or truncated

            # Apply reward modifier if specified
            if reward_modifier is not None:
                reward = reward_modifier(state, action, reward, done)
                #print(f"Modified Reward: {reward}, Original Reward: {env.step(action)[1]}")

            # Incremental reward tracking
            total_reward += reward

            # Compute features for the next state
            next_features = rbf.encode(next_state)

            # Choose next action using epsilon-greedy policy
            if np.random.rand() < epsilon:
                next_action = env.action_space.sample()
            else:
                q_values_next = np.dot(weights, next_features)
                next_action = np.argmax(q_values_next)

            # SARSA update
            q_value = np.dot(weights[action], features)
            next_q_value = np.dot(weights[next_action], next_features) if not done else 0
            td_error = reward + gamma * next_q_value - q_value
            weights[action] += alpha * td_error * features

            # Transition to the next state and action
            state, action, features = next_state, next_action, next_features

        # Decay epsilon for exploration
        epsilon = max(epsilon_min, epsilon * decay_rate)

        # Track total rewards for the episode
        rewards_per_episode.append(total_reward)

        if ep % 10 == 0:
            avg_reward = np.mean(rewards_per_episode[-10:])
            print(f"Episode {ep}, Avg Reward: {avg_reward:.2f}, Epsilon: {epsilon:.4f}")

    return weights, rewards_per_episode, rbf


def cartpole_reward_modifier(state, action, reward, step_count):
    # Extract state variables
    position = state[0]  # Cart position
    velocity = state[1]  # Cart velocity
    pole_angle = state[2]  # Pole angle
    pole_velocity = state[3]  # Pole angular velocity

    # Bonus for stability (small pole angle and angular velocity)
    if abs(pole_angle) < 0.05 and abs(pole_velocity) < 0.1:
        reward += 3  # Stability bonus

    # Penalty for instability (large pole angle or velocity)
    else:
        reward -= 2

    # Additional penalty for large cart displacement
    if abs(position) > 1.0:  # Penalize large displacement from center
        reward -= 1

    # Scaling survival reward (reward grows with step count)
    reward += step_count * 0.1

    return reward

def mountaincar_reward_modifier(state, action, reward, done):
    # Extract state variables
    position = state[0]  # Car position
    velocity = state[1]  # Car velocity

    # Scale reward continuously based on position
    normalized_position = (position + 1.2) / 1.7  # Scale position to [0, 1]
    reward += 100 * normalized_position  # Reward for moving right

    # Reward for positive velocity (rightward movement)
    if velocity > 0:
        reward += 5 * velocity  # Reward scaled by velocity magnitude

    # Penalize for negative velocity (leftward movement)
    if velocity < 0:
        reward -= 2 * abs(velocity)

    # Success bonus for reaching the goal
    if done and position >= 0.5:
        reward += 1000  # Large bonus for success

    return reward

def lunarlander_reward_modifier(state, action, reward, done):
    # Extract state variables
    horizontal_position = state[0]
    vertical_position = state[1]
    horizontal_velocity = state[2]
    vertical_velocity = state[3]
    angle = state[4]  # Angle of the lander
    angular_velocity = state[5]  # Angular velocity

    if abs(horizontal_velocity) < 0.1 and abs(angle) < 0.1:
        reward += 20  # Stability bonus


    return reward
```

```python
class FourierBasis:
    def __init__(self, state_dim, order, low, high):
        self.state_dim = state_dim
        self.order = order
        self.low = np.array(low)
        self.high = np.array(high)
        # Create coefficients for the Fourier basis
        self.coefficients = np.indices([order + 1] * state_dim).reshape(state_dim, -1).T

    def encode(self, state):
        # Normalize state to [0, 1]
        scaled_state = (state - self.low) / (self.high - self.low)
        return np.cos(np.pi * np.dot(self.coefficients, scaled_state))

def semi_gradient_qlearning(
    env_name,
    episodes=1000,
    gamma=0.95,
    alpha=0.01,
    epsilon=0.3,
    epsilon_min=0.05,
    decay_rate=0.999,
    order=3,  # Order of the Fourier basis
    reward_modifier=None  # Optional reward adjustment function
):
    env = gym.make(env_name)
    state_dim = env.observation_space.shape[0]
    action_dim = env.action_space.n

    low = env.observation_space.low
    high = env.observation_space.high
    fourier = FourierBasis(state_dim, order, low, high)
    feature_dim = fourier.coefficients.shape[0]
    weights = np.random.normal(0, 0.1, (action_dim, feature_dim))

    rewards_per_episode = []

    for ep in range(episodes):
        state = env.reset(seed=42)[0]  # Updated for compatibility
        total_reward = 0
        done = False

        while not done:
            features = fourier.encode(state).flatten()  # Ensure features are 1D

            # Epsilon-greedy policy
            if np.random.rand() < epsilon:
                action = env.action_space.sample()
            else:
                q_values = np.dot(weights, features)
                action = np.argmax(q_values)

            # Take a step
            next_state, reward, done, _ = env.step(action)

            if reward_modifier is not None:
                reward = reward_modifier(state, action, reward, done)

            reward = np.clip(reward, -100, 100)
            total_reward += reward

            next_features = fourier.encode(next_state).flatten()  # Ensure features are 1D

            # TD Target
            max_next_q = np.max(np.dot(weights, next_features)) if not done else 0
            td_target = reward + gamma * max_next_q

            # TD Error
            q_value = np.dot(weights[action], features)
            td_error = td_target - q_value

            # Update weights
            weights[action] += alpha * td_error * features

            state = next_state

        # Decay epsilon
        epsilon = max(epsilon_min, epsilon * decay_rate)
        rewards_per_episode.append(total_reward)

        if ep % 10 == 0:
            avg_reward = np.mean(rewards_per_episode[-10:])
            print(f"Episode {ep}, Avg Reward: {avg_reward:.2f}, Epsilon: {epsilon:.4f}")

    return weights, rewards_per_episode, fourier

def playback_episode(env_name, trained_weights, fourier, max_step):
    # Wrap the environment for playback
    env = wrap_env(
        gym.make(env_name, render_mode="rgb_array", new_step_api=True),
        video_folder='./video',
        run_id=3
    )
    env = TimeLimit(env, max_episode_steps=max_step)

    # Reset environment
    state = env.reset(seed=42)[0]

    done = False
    step_count = 0
    total_reward = 0

    print("Starting playback of an episode...")

    while step_count < max_step and not done:
        # Compute action using the trained weights
        features = fourier.encode(state).flatten()  # Ensure 1D features
        q_values = np.dot(trained_weights, features)
        action = np.argmax(q_values)

        # Take a step in the environment
        next_state, reward, done, _ = env.step(action)

        total_reward += reward
        step_count += 1

        # Print playback details
        print(f"Step {step_count}: State={state}, Action={action}, Reward={reward}, Total={total_reward}")

        state = next_state

    print(f"Playback complete. Total reward for this episode: {total_reward:.2f}")

    env.close()
    show_video(f'./video/run_3')
```
