```python
import cv2
import numpy as np
from concurrent.futures import ThreadPoolExecutor

def preprocess_frames(frames):
    # Convert all frames to grayscale
    gray_frames = np.dot(frames[...,:3], [0.2989, 0.5870, 0.1140])  # RGB to Grayscale

    # Resize frames to 84x84
    resized_frames = np.array([cv2.resize(frame, (84, 84)) for frame in gray_frames])

    # Normalize pixel values and flatten
    processed_frames = resized_frames.reshape(resized_frames.shape[0], -1) / 255.0
    return processed_frames


def preprocess_frame(frame):
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
    frame = cv2.resize(frame, (84, 84))
    frame = frame.flatten() / 255.0
    return frame

def preprocess_frames_parallel(frames):
    with ThreadPoolExecutor() as executor:
        processed_frames = list(executor.map(preprocess_frame, frames))
    return np.array(processed_frames)

def get_positions(frame):
    """
    Extracts the center positions of the left paddle, right paddle, and the ball.

    Args:
        frame (numpy.ndarray): The RGB frame from the game environment.

    Returns:
        left_paddle_center (float): The vertical center position (y-coordinate) of the left paddle.
        right_paddle_center (float): The vertical center position (y-coordinate) of the right paddle.
        ball_center (tuple): The (x, y) center position of the ball.
    """
    # Convert to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
    height, width = gray_frame.shape

    # Define ROIs for paddles
    left_paddle_roi = gray_frame[:, :width // 4]  # Left quarter
    right_paddle_roi = gray_frame[:, 3 * width // 4:]  # Right quarter

    # Threshold and detect contours for paddles
    _, left_thresh = cv2.threshold(left_paddle_roi, 200, 255, cv2.THRESH_BINARY)
    _, right_thresh = cv2.threshold(right_paddle_roi, 200, 255, cv2.THRESH_BINARY)

    left_contours, _ = cv2.findContours(left_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    right_contours, _ = cv2.findContours(right_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # Threshold and detect contours for the ball (entire frame)
    _, ball_thresh = cv2.threshold(gray_frame, 200, 255, cv2.THRESH_BINARY)
    ball_contours, _ = cv2.findContours(ball_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # Helper function to get the center of the largest contour
    def get_largest_contour_center(contours, roi_offset_x=0):
        if len(contours) == 0:
            return None  # No object detected
        largest_contour = max(contours, key=cv2.contourArea)
        x, y, w, h = cv2.boundingRect(largest_contour)
        center_x = x + w // 2 + roi_offset_x
        center_y = y + h // 2
        return (center_x, center_y)

    # Get paddle and ball centers
    left_paddle_center = get_largest_contour_center(left_contours, roi_offset_x=0)
    right_paddle_center = get_largest_contour_center(right_contours, roi_offset_x=3 * width // 4)
    ball_center = get_largest_contour_center(ball_contours)

    # For paddles, return only the vertical center (y-coordinate)
    left_paddle_y = left_paddle_center[1] if left_paddle_center else None
    right_paddle_y = right_paddle_center[1] if right_paddle_center else None
    ball_xy = ball_center if ball_center else (None, None)

    return left_paddle_y, right_paddle_y, ball_xy

def reset_paddle_to_center(frame, paddle_side="right"):
    """
    Resets the paddle position to the center of the game board in the provided frame.
    :param frame: The current rendered frame from the environment (RGB array).
    :param paddle_side: Which paddle to reset ("left" or "right").
    :return: Modified frame with the paddle reset.
    """
    height, width, _ = frame.shape
    paddle_center_y = height // 2  # Center of the screen

    if paddle_side == "right":
        x_start = width - 15  # Right paddle's x-coordinate
        x_end = width - 10
    elif paddle_side == "left":
        x_start = 10
        x_end = 15

    # Clear existing paddle (black it out)
    frame[:, x_start:x_end] = 0

    # Draw paddle at center
    frame[paddle_center_y - 15:paddle_center_y + 15, x_start:x_end] = [255, 255, 255]  # White paddle
    return frame


def infer_score(ball_xy, screen_width):
    """
    Infers if a point was scored based on the ball position.
    :param ball_xy: Tuple of ball coordinates (x, y).
    :param screen_width: Width of the game screen.
    :return: True if a point was scored, else False.
    """
    if ball_xy[0] < 0:  # Ball passed the left edge
        return "right"
    elif ball_xy[0] > screen_width:  # Ball passed the right edge
        return "left"
    return None

```

```python
from collections import deque

class PrioritizedReplayBuffer:
    def __init__(self, capacity):
        self.capacity = capacity
        self.buffer = []
        self.priorities = []

    def add(self, transition, priority):
        if len(self.buffer) < self.capacity:
            self.buffer.append(transition)
            self.priorities.append(priority)
        else:
            self.buffer.pop(0)
            self.priorities.pop(0)
            self.buffer.append(transition)
            self.priorities.append(priority)

    def sample(self, batch_size):
        # Normalize priorities
        scaled_priorities = np.array(self.priorities) + 1e-6
        probabilities = scaled_priorities / np.sum(scaled_priorities)
        indices = np.random.choice(len(self.buffer), batch_size, p=probabilities)
        samples = [self.buffer[i] for i in indices]
        return samples, indices

    def update_priorities(self, indices, new_priorities):
        for i, idx in enumerate(indices):
            self.priorities[idx] = new_priorities[i]
```

```python
from tensorflow.keras import models, layers, optimizers

# Neural Network for Q-value approximation
def create_q_network(input_dim, action_dim):
    model = models.Sequential([
        layers.Dense(256, activation='relu', input_shape=(input_dim,)),
        layers.Dense(256, activation='relu'),
        layers.Dense(128, activation='relu'),
        layers.Dense(action_dim)
    ])

    model.compile(optimizer=optimizers.Adam(learning_rate=0.001), loss='mse')
    return model
```

```python
import gym

# Environment setup
env = gym.make('ALE/Pong-v5', render_mode="rgb_array")
action_dim = env.action_space.n
input_dim = 84 * 84  # Flattened grayscale frame dimensions

# Initialize Q-networks
q_network_1 = create_q_network(input_dim, action_dim)
q_network_2 = create_q_network(input_dim, action_dim)

# Initialize replay buffer
replay_buffer = PrioritizedReplayBuffer(capacity=10000)
```

```python
BATCH_SIZE = 64
for _ in range(BATCH_SIZE * 20):  # Increase pretraining size
    raw_state = env.reset()
    state = preprocess_frame(raw_state)
    while True:
        action = env.action_space.sample()
        raw_next_state, reward, done, _ = env.step(action)
        next_state = preprocess_frame(raw_next_state)
        td_error = 1.0
        replay_buffer.add((state, action, reward, next_state, done), td_error)
        state = next_state
        if done:
            break
action_counts = {i: 0 for i in range(env.action_space.n)}
```

```python
def record_episode(env, model, video_path, episode):
    frames = []
    state = preprocess_frame(env.render(mode="rgb_array"))

    while True:
        # Predict the best action using the trained model
        q_values = model.predict(np.expand_dims(state, axis=0))
        action = np.argmax(q_values[0])

        # Take the action
        raw_next_state, reward, done, _ = env.step(action)

        # Append the current frame to the video
        frames.append(env.render(mode="rgb_array"))
        state = preprocess_frame(raw_next_state)

        if done:
            print(f"Episode {episode + 1} finished during recording.")
            break

    # Save the frames to a video
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    height, width, _ = frames[0].shape
    video = cv2.VideoWriter(video_path, fourcc, 30, (width, height))

    for frame in frames:
        video.write(cv2.cvtColor(frame, cv2.COLOR_RGB2BGR))

    video.release()
    print(f"Video for Episode {episode + 1} saved to {video_path}")
```

```python
import tensorflow as tf
import numpy as np
import cv2
from google.colab import files

# Hyperparameters
GAMMA = 0.99
NUM_EPISODES = 2
EPSILON = 1.0
EPSILON_DECAY = 0.995
MIN_EPSILON = 0.1
UPDATE_TARGET_STEPS = 50
BATCH_SIZE = 64

def record_episode(env, model, video_path, episode):
    """Record a full episode using the trained model."""
    raw_state = env.reset()
    state = preprocess_frame(raw_state)
    frames = []
    total_reward = 0
    action_counts = {0: 0, 1: 0, 2: 0, 3: 0}

    while True:
        q_values = model.predict(np.expand_dims(state, axis=0))
        action = np.argmax(q_values[0])
        action_counts[action] += 1

        raw_next_state, reward, done, _ = env.step(action)
        total_reward += reward

        frames.append(env.render(mode="rgb_array"))
        state = preprocess_frame(raw_next_state)

        if done:
            break

    # Save the video
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    height, width, _ = frames[0].shape
    video = cv2.VideoWriter(video_path, fourcc, 30, (width, height))

    for frame in frames:
        video.write(cv2.cvtColor(frame, cv2.COLOR_RGB2BGR))

    video.release()
    print(f"Episode {episode + 1} recorded: Total Reward = {total_reward}, Actions = {action_counts}")
    files.download(video_path)

# Training loop
rewards_per_episode = []
with tf.device('/GPU:0'):
    for episode in range(NUM_EPISODES):
        raw_state = env.reset()
        state = preprocess_frame(raw_state)
        total_reward = 0
        action_counts = {0: 0, 1: 0, 2: 0, 3: 0}

        while True:
            if np.random.rand() < EPSILON:
                action = np.random.choice([2, 3])  # Randomly choose valid actions (RIGHT or LEFT)
            else:
                q_values = q_network_1.predict(np.expand_dims(state, axis=0))
                action = np.argmax(q_values[0])
                if action not in [2, 3]:
                    action = np.random.choice([2, 3])

            action_counts[action] += 1
            raw_next_state, reward, done, _ = env.step(action)
            next_state = preprocess_frame(raw_next_state)

            # Reward shaping
            _, next_right_paddle_y, next_ball_xy = get_positions(raw_next_state)
            _, right_paddle_y, ball_xy = get_positions(raw_state)

            if ball_xy and right_paddle_y and next_ball_xy and next_right_paddle_y:
                prev_distance = abs(ball_xy[1] - right_paddle_y)
                next_distance = abs(next_ball_xy[1] - next_right_paddle_y)
                reward += 0.1 if next_distance < prev_distance else -0.1

            total_reward += reward

            # Compute TD-error
            q_values = q_network_1.predict(np.expand_dims(state, axis=0))
            next_q_values_1 = q_network_1.predict(np.expand_dims(next_state, axis=0))
            next_q_values_2 = q_network_2.predict(np.expand_dims(next_state, axis=0))

            best_action = np.argmax(next_q_values_1[0])
            td_target = reward + (1 - done) * GAMMA * next_q_values_2[0][best_action]
            td_error = abs(td_target - q_values[0][action])

            replay_buffer.add((state, action, reward, next_state, done), td_error)
            state = next_state

            # Train the networks if replay buffer has enough samples
            if len(replay_buffer.buffer) >= BATCH_SIZE:
                samples, indices = replay_buffer.sample(BATCH_SIZE)
                states, actions, rewards, next_states, dones = zip(*samples)
                states = tf.convert_to_tensor(np.vstack(states), dtype=tf.float32)
                next_states = tf.convert_to_tensor(np.vstack(next_states), dtype=tf.float32)

                targets_1 = q_network_1.predict(states)
                next_q_values_1 = q_network_1.predict(next_states)
                next_q_values_2 = q_network_2.predict(next_states)

                for i, (state, action, reward, next_state, done) in enumerate(samples):
                    best_action = np.argmax(next_q_values_1[i])
                    td_target = reward + (1 - done) * GAMMA * next_q_values_2[i][best_action]
                    targets_1[i][action] = td_target

                q_network_1.fit(states, targets_1, epochs=1, verbose=0, batch_size=BATCH_SIZE)
                updated_q_values = q_network_1.predict(states)
                td_errors = [
                    abs(reward + (1 - done) * GAMMA * np.max(next_q_values_2[i]) - updated_q_values[i][action])
                    for i, (state, action, reward, next_state, done) in enumerate(samples)
                ]
                replay_buffer.update_priorities(indices, td_errors)

            if done:
                break

        # Logging
        print(f"Episode {episode + 1}/{NUM_EPISODES}, Reward: {total_reward}, Actions: {action_counts}")
        rewards_per_episode.append(total_reward)

        # Record video for this episode
        video_path = f"/content/pong_episode_{episode + 1}.mp4"
        record_episode(env, q_network_1, video_path, episode)

        # Update target network periodically
        if episode % UPDATE_TARGET_STEPS == 0:
            q_network_2.set_weights(q_network_1.get_weights())

        # Decay epsilon
        EPSILON = max(MIN_EPSILON, EPSILON * EPSILON_DECAY)

        # Save model weights
        q_network_1.save(f"q_network_1_episode_{episode}.keras")
        q_network_2.save(f"q_network_2_episode_{episode}.keras")
        print(f"Model saved at Episode {episode}")
```

```python
def evaluate_agent(env, model, num_games=100):
    total_steps = 0
    total_score = 0

    for game in range(num_games):
        #raw_state = env.reset()
        state = preprocess_frame(raw_state)
        steps = 0
        score = 0

        # Initialize action counts for this game
        action_counts = {0: 0, 1: 0, 2: 0, 3: 0}

        while True:
            # Predict the best action using the model
            q_values = model.predict(np.expand_dims(state, axis=0))
            action = np.argmax(q_values[0])

            # Count the action
            if action in action_counts:
                action_counts[action] += 1

            # Take the action in the environment
            raw_next_state, reward, done, _ = env.step(action)
            state = preprocess_frame(raw_next_state)
            steps += 1
            score += reward

            if done:
                break

        total_steps += steps
        total_score += score

        # Print action counts for this game
        print(f"Game {game + 1}/{num_games}: Total Steps = {steps}, Score = {score}")
        print(f"Action counts: {action_counts}")

    avg_steps = total_steps / num_games
    avg_score = total_score / num_games
    return avg_steps, avg_score
```
